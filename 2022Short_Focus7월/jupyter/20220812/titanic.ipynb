```python
def age_change(x):
    if x < 10:
        return 1
    elif x < 20:
        return 10
    elif x < 30:
        return 20
    elif x < 40:
        return 30
    elif x < 50:
        return 40
    elif x < 60:
        return 50
    elif x < 70:
        return 60
    elif x < 80:
        return 70
    else :
        return 80
```

```python
age_change_list= []
for i in data['Age']:
    age_change_list.append(age_change(i))
age_change_list
```

```python
data['age_change'] = age_change_list
data.head()
```

```python
y = data.iloc[:, 1]
y
```

```python
X = data.drop(['age_change', 'Died', 'Survived'], axis=1)
X
```

```python
X = X.drop(['Name'], axis=1)
X
```

```python
X = X.drop(['Cabin'], axis=1)
X
```

```python
X['Embarked'].value_counts()
```

```python
X['Embarked'] = X['Embarked'].fillna('S')
```

```python
X.info()
```

```python
X
```

```python
dic = {'male':1, 'female':0}
X['Sex'] = X['Sex'].map(dic)
X
```

```python
pclass_dummies = pd.get_dummies(X['Pclass'], prefix='Pclass')
pclass_dummies
```

```python
X = pd.concat([X, pclass_dummies], axis=1)
X
```

```python
X.drop('Pclass', axis=1, inplace=True)
X
```

```python
X.drop('Ticket', axis=1, inplace=True)
X
```

```python
embarked_dummies = pd.get_dummies(X['Embarked'], prefix='Embarked')
embarked_dummies
```

```python
X = pd.concat([X, embarked_dummies], axis=1)
```

```python
X
```

```python
X.drop('Embarked', axis=1, inplace=True)
```

```python
X
```

```python
X.drop('PassengerId', axis=1, inplace=True)
X
```

```python
y
```

```python
X_train, y_train, X_test, y_test = train_test_split(X, y, test_size=0.3)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
score_list = []
```

```python
for i in range(1, 101):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train, X_test)
    pre = model.predict(y_train)
    score = metrics.accuracy_score(y_test, pre)
    score_list.append(score)
score_list
```

```python
index = 0
max = score_list[0]
for k in range(0, 100):
    if max < score_list[k]:
        max = score_list[k]
max
```

```python
print('Accuract : {:.2f}%'.format(score*100))
```

```python
modelTree = DecisionTreeClassifier()
```

```python
modelTree.fit(X_train, X_test)
```

```python
pre = modelTree.predict(y_train)
```
