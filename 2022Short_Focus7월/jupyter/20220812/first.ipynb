```python
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = pd.read_csv('./data/mushroom.csv')
data
```

```python
data.describe()
```

# Data split

```python
X = data.iloc[:, 1:]
y = data.iloc[:, 0]
X.shape
```

```python
y.shape
```

```python
target_dic = {'p':0, 'e':1}
y = y.map(target_dic)
y
```

```python
pd.get_dummies(X.loc[:, 'cap-shape'])
```

```python
pd.get_dummies(data.loc[:, "cap-shape"], prefix='cap-shape')
```

```python
X
```

```python
X = pd.get_dummies(X, prefix=X.columns)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = KNeighborsClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
pre = model.predict(X_test)
```

```python
pre
```

```python
from sklearn import metrics
```

```python
score = metrics.accuracy_score(y_test, pre)
```

```python
score
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
modelTree = DecisionTreeClassifier(max_depth=7)
```

```python
modelTree.fit(X_train, y_train)
```

```python
scoreTree = modelTree.score(X_test, y_test)
scoreTree
```

```python
modelTree.feature_importances_
```

```python
import_list = list(zip(modelTree.feature_importances_, X.columns))
import_list.sort()
import_list.reverse()
import_list[:5]
```
